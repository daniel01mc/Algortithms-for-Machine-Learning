### 6. Batch Gradient Descent
---

**I.** Implement Batch Gradient Descent based on the way we have learned in the class (See sample code form pdf). You can play with eta and n_iterations and should set to reasonable eta and number of iterations so that you can get the thetas close to Normal equation's theta

```python

y = np.array(y_train)


cost_list = []
epoch_list = []
predicted_list = []


m= 100

n_iterations = 1000
eta = 0.1
theta_gd = np.random.randn(5,1)  # random initialization



y = y.reshape(-1,1)

print(y.shape)

for iteration in range(n_iterations):
    gradients = 2/m * x_a.T.dot(x_a.dot(theta_gd) - y)
    theta_gd = theta_gd - eta * gradients

    y_predicted = np.dot(theta_gd.T, x_b.T)
    cost = np.mean(np.square(y-y_predicted)) # MSE (Mean Squared Error)
        
    
    cost_list.append(cost)
    epoch_list.append(iteration)
```

### 12. Learning Curve
---

**I.** Generate learning curve with linearRegression

**II.** Generate learning curve with polynomial regression with degree  = 5

**III.** Interpret the result

```python
def plot_learning_curves_lineal(model, X, y):

  X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
  train_errors, val_errors = [], []

  fig, ax = plt.subplots()
  for m in range(1, len(X_train)):

    model.fit(X_train[:m], y_train[:m])
    y_train_predict = model.predict(X_train[:m])
    y_val_predict = model.predict(X_val)
    train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
    val_errors.append(mean_squared_error(y_val, y_val_predict))

  ax.plot(np.sqrt(train_errors), "r", label="train")
  ax.plot(np.sqrt(val_errors), "b:", label="val")
    
  ax.legend( shadow=True)
  plt.ylabel("RMSE")
  plt.xlabel("Training set size") 
  ax.set_ylim([0,20])
  
  plt.show()  

lin_reg = LinearRegression()
plot_learning_curves_lineal(lin_reg, X,Y)
```

```python
from sklearn.pipeline import Pipeline


def plot_learning_curves_poly(model, X, y):

  X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
  train_errors, val_errors = [], []

  fig, ax = plt.subplots()
  for m in range(1, len(X_train)):

    model.fit(X_train[:m], y_train[:m])
    y_train_predict = model.predict(X_train[:m])
    y_val_predict = model.predict(X_val)
    train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
    val_errors.append(mean_squared_error(y_val, y_val_predict))

  ax.plot(np.sqrt(train_errors), "r", label="train")
  ax.plot(np.sqrt(val_errors), "b:", label="val")
    
  ax.legend( shadow=True)
  plt.ylabel("RMSE")
  plt.xlabel("Training set size") 
  ax.set_ylim([0,100])
  
  
  plt.show()  


polynomial_regression = Pipeline([("poly_features", PolynomialFeatures(degree=5,include_bias=False)), ("lin_reg", LinearRegression())])
plot_learning_curves_poly(polynomial_regression, X, Y)
```

**Interpreting the results:** By using the polynomial learning curve the error for our validation set is much bigger compared to the one using the linear regression approach. Therefore, the model is overfitting
